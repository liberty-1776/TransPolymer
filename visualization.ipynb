```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pretrain_df = pd.read_csv('mywork/result_data/Non-Augmented/Ei/PCA_E_pretrain_fingerprint.csv')
best_model_df = pd.read_csv('mywork/result_data/Non-Augmented/Egc/PCA_Egc_best_fingerprint.csv')
```

```python
pretrain_pc1 = pretrain_df['PC1']
pretrain_pc2 = pretrain_df['PC2']
pretrain_orig_out = pretrain_df['orig_out']
best_model_pc1 = best_model_df['PC1']
best_model_pc2 = best_model_df['PC2']
best_model_orig_out = best_model_df['orig_out']
```

```python
#plt.scatter(pretrain_pc1,pretrain_pc2, c=pretrain_orig_out, cmap='cool',s=2,alpha=0.6, label='Pretrain Fingerprint')
plt.scatter(best_model_pc1,best_model_pc2, c=best_model_orig_out,s=2,alpha=0.6, cmap='hot', label='Best Fingerprint')

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Scatter Plot of PC1 and PC2(Egc) for best_model (Color: orig_out)')
plt.legend()

# Show the colorbar
plt.colorbar(label='orig_out')
# Show the plot
plt.savefig('mywork/figs/Egc/pca_Egc_best.png')
plt.show()
plt.close()
```
