```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pretrain_df = pd.read_csv('mywork/result_data/Ei/PCA_Ei_pretrain_fingerprint_unscaled.csv')
best_model_df = pd.read_csv('mywork/result_data/Ei/PCA_Ei_best_fingerprint_unscaled.csv')
```

```python
pretrain_pc1 = pretrain_df['PC1']
pretrain_pc2 = pretrain_df['PC2']
pretrain_orig_out = pretrain_df['orig_out']
best_model_pc1 = best_model_df['PC1']
best_model_pc2 = best_model_df['PC2']
best_model_orig_out = best_model_df['orig_out']
```

```python
plt.scatter(pretrain_pc1,pretrain_pc2, c=pretrain_orig_out, cmap='cool', label='Pretrain Fingerprint')
#plt.scatter(best_model_pc1,best_model_pc2, c=best_model_orig_out, cmap='hot', label='Best Fingerprint')

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Scatter Plot of PC1 and PC2(Ei) for Best_model (Color: orig_out)')
plt.legend()

# Show the colorbar
plt.colorbar(label='orig_out')
plt.savefig('mywork/figs/Ei/pca_pretrain_visualization.png')
# Show the plot
plt.show()
plt.close()
```
