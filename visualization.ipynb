```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pretrain_df = pd.read_csv('mywork/result_data/Non-Augmented/Eea/TSNE_Eea_pretrain_fingerprint.csv')
best_model_df = pd.read_csv('mywork/result_data/Non-Augmented/Eea/TSNE_Eea_best_fingerprint.csv')
```

```python
pretrain_pc1 = pretrain_df['tsne_1']
pretrain_pc2 = pretrain_df['tsne_2']
pretrain_orig_out = pretrain_df['orig_out']
best_model_pc1 = best_model_df['tsne_1']
best_model_pc2 = best_model_df['tsne_2']
best_model_orig_out = best_model_df['orig_out']
```

```python
plt.scatter(pretrain_pc1,pretrain_pc2, c=pretrain_orig_out, cmap='cool',s=5,alpha=0.8, label='Pretrain Fingerprint')
#plt.scatter(best_model_pc1,best_model_pc2, c=best_model_orig_out,s=5,alpha=0.8, cmap='hot', label='Best Fingerprint')

plt.xlabel('tsne_1')
plt.ylabel('tsne_2')
plt.title('Scatter Plot of tsne_1 and tsne_2(Eea) for pretrain_model (Color: orig_out)')
plt.legend()

# Show the colorbar
plt.colorbar(label='orig_out')
# Show the plot
plt.show()
plt.savefig('mywork/figs/Eea/tsne_Eea_pretrain.png')
```
