```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pretrain_df = pd.read_csv('mywork/reduced_pretrain_Egc_scaled.csv')
best_model_df = pd.read_csv('mywork/reduced_Egc_best_fingerprint_scaled.csv')
```

```python
pretrain_pc1 = pretrain_df['PC1']
pretrain_pc2 = pretrain_df['PC2']
best_model_pc1 = best_model_df['PC1']
best_model_pc2 = best_model_df['PC2']
```

```python
plt.scatter(pretrain_pc1, pretrain_pc2, c='blue', label='Pretrain Fingerprint')
plt.scatter(best_model_pc1, best_model_pc2, c='red', label='Best Model Fingerprint')

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Scatter Plot of PC1 and PC2')
plt.legend()

plt.show()
```
