```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
best_model_df = pd.read_csv('mywork/result_data/Augmented/Egb/TSNE_Egb_best_fingerprint.csv')
```

```python
best_model_pc1 = best_model_df['tsne_1']
best_model_pc2 = best_model_df['tsne_2']
best_model_orig_out = best_model_df['orig_out']
```

```python
plt.scatter(best_model_pc1,best_model_pc2, c=best_model_orig_out,s=5,alpha=0.8, cmap='hot', label='Best Fingerprint')

plt.xlabel('tsne_1')
plt.ylabel('tsne_2')
plt.title('Scatter Plot of tsne_1 and tsne_2(Egb) for best_model (Color: orig_out)')
plt.legend()

# Show the colorbar
plt.colorbar(label='orig_out')
# Show the plot
plt.savefig('mywork/figs/Augmented/Egb/tsne_Eea_best.png')
plt.show()
plt.close()
```
